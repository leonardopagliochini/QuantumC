```python
# ðŸš€ Configuration: set number of bits and drawing mode
NUM_BITS = 2
DRAW_MODE = 'mpl'  # Options: 'mpl', 'text', 'latex', 'latex_source'

# ðŸ§  Load arithmetic module
from qiskit import QuantumCircuit
import matplotlib.pyplot as plt
import sys
sys.path.append(".")
from step5_quantum_mlir_to_qasm import q_arithmetics as qutil
qutil.set_number_of_bits(NUM_BITS)

# Helper: safe values
MIN_VAL = -(2 ** (NUM_BITS - 1))
MAX_VAL =  (2 ** (NUM_BITS - 1)) - 1
```

```python
# addi_in_place
qc = QuantumCircuit()
areg = qutil.initialize_variable(qc, MAX_VAL - 1)
qutil.addi_in_place(qc, areg, -1)
qc.draw(DRAW_MODE)
```

```python
# addi
qc = QuantumCircuit()
areg = qutil.initialize_variable(qc, 1)
qutil.addi(qc, areg, -1)
qc.draw(DRAW_MODE)
```

```python
# add_in_place
qc = QuantumCircuit()
a = qutil.initialize_variable(qc, 1)
b = qutil.initialize_variable(qc, 1)
qutil.add_in_place(qc, a, b)
qc.draw(DRAW_MODE)
```

```python
# subi
qc = QuantumCircuit()
a = qutil.initialize_variable(qc, MAX_VAL)
qutil.subi(qc, a, 1)
qc.draw(DRAW_MODE)
```
